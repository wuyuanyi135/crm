```python
from tests.get_data import get_sample_data, get_sample_data_polymorphic
from crm.presets.hypothetical import Hypothetical1D, HypotheticalPolymorphic1D
import pandas as pd
import numpy as np
```

```python
# data = get_sample_data()
# system_spec = Hypothetical1D()
# form_names = system_spec.get_form_names()
data = get_sample_data_polymorphic()
system_spec = HypotheticalPolymorphic1D()
form_names = system_spec.get_form_names()
```

```python
d = [x.__dict__ for x in data]
```

```python
df = pd.DataFrame.from_records(d).set_index("time")
time = df.index
```

```python
df_ct = df[["concentration", "temperature"]]
df_ct
```

```python
n_column_index = pd.MultiIndex.from_product([("n",), form_names])
df_n = pd.DataFrame(df["n"].tolist(), columns=n_column_index, index=time)
df_n
```

```python
df_extra = pd.DataFrame(df["extra"].tolist(), index=time)
df_extra
```

```python
vf_index = pd.MultiIndex.from_product([("vf", ), form_names])
df_vf = pd.DataFrame(df_extra["vfs"].tolist(), columns=vf_index, index=time)
df_vf
```

```python
gds_col_index = pd.MultiIndex.from_product([("gds", ), form_names])
df_gds = pd.DataFrame(df_extra["gds"].tolist(), columns=gds_col_index, index=time)
df_gds
```

```python
nuc_col_index = pd.MultiIndex.from_product([form_names, ("pn", "sn")])
df_nuc = pd.DataFrame(np.array(df_extra["nucleation"].values.tolist()).reshape((-1, 2*len(form_names))), index=time, columns=nuc_col_index).swaplevel(axis=1).sort_index(1)
df_nuc
```

```python
df_profiling = pd.DataFrame.from_dict(df_extra["profiling"].tolist())
df_profiling.index = time
df_profiling
```

```python
sol_ss_col_index = pd.MultiIndex.from_product([("sol", "ss"), form_names])
df_sol_ss = pd.DataFrame(np.array(df_extra[["sols", "sses"]].values.tolist()).reshape(-1, 2*len(form_names)), columns=sol_ss_col_index, index=time)
df_sol_ss
```

```python
df_summary = pd.concat([df_ct, df_n, df_gds, df_nuc], axis=1)
df_summary
```
